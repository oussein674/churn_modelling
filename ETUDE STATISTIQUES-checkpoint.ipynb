```python



import pandas as pd

# Charger le fichier de données
file_path = 'Documents\\Churn_Modelling.csv'
data = pd.read_csv(file_path)

# Initialiser une liste pour stocker les informations des colonnes à analyser
columns_info = []

# Parcourir chaque colonne pour obtenir les informations requises, en excluant 'CustomerId' et 'Surname'
for column in data.columns:
    if column not in ['CustomerId', 'Surname']:  # Exclure les colonnes non analysées
        # Nom de la colonne
        column_name = column
        # Description par défaut (modifiable si nécessaire)
        description = f"Description de la variable '{column}'"
        # Type de données
        dtype = data[column].dtype
        # Modalités (uniquement pour les colonnes catégorielles ou ayant peu de valeurs uniques)
        if dtype == 'object' or data[column].nunique() < 20:
            modalities = data[column].unique().tolist()
        else:
            modalities = 'N/A'
        
        # Ajouter les informations de la colonne au tableau
        columns_info.append([column_name, description, dtype, modalities])

# Créer un DataFrame contenant les informations
columns_df = pd.DataFrame(columns_info, columns=['Nom de la variable', 'Description', 'Type', 'Modalités'])

# Appliquer un style au tableau pour un affichage coloré
styled_columns_df = columns_df.style.set_properties(**{
    'background-color': '#e1f7d5',  # Couleur de fond (choix de vert clair)
    'color': '#333333',             # Couleur du texte
    'border-color': 'black',        # Couleur de bordure
    'border-style': 'solid',        # Style de bordure
    'border-width': '1px',          # Épaisseur de bordure
    'padding': '8px'                # Espacement intérieur
}).set_table_styles([
    {'selector': 'thead th', 'props': [('background-color', '#b2df8a'), ('color', 'black'), ('border-color', 'black'), ('border-style', 'solid'), ('border-width', '1px')]}
]).set_caption("Tableau d'analyse des variables")

# Afficher le tableau stylisé
styled_columns_df


```

# tableau de la variable EstimatedSalary

```python
import pandas as pd

# Charger les données
data = pd.read_csv('Documents\\Churn_Modelling.csv')

# Séparer les clients en deux groupes : ceux qui sont partis (Exited=1) et ceux qui ne sont pas partis (Exited=0)
partis = data[data['Exited'] == 1]['EstimatedSalary']
non_partis = data[data['Exited'] == 0]['EstimatedSalary']

# Calcul des statistiques pour chaque groupe
statistiques = {
    'Statistiques': ['Q1', 'Q2 (Médiane)', 'Q3', 'Q99'],
    'Partis': [
        partis.quantile(0.25),  # Q1
        partis.median(),        # Q2 (médiane)
        partis.quantile(0.75),  # Q3
        partis.quantile(0.99)   # Q99
    ],
    'Non partis': [
        non_partis.quantile(0.25),  # Q1
        non_partis.median(),        # Q2 (médiane)
        non_partis.quantile(0.75),  # Q3
        non_partis.quantile(0.99)   # Q99
    ]
}

# Créer un tableau pour présenter les résultats
resultats = pd.DataFrame(statistiques)

# Afficher le tableau
print(resultats)


```

# Tableau de la variable age

```python
import pandas as pd

# Chargement des données
file_path = 'Documents\\Churn_Modelling.csv'
data = pd.read_csv(file_path)

# Calcul des statistiques descriptives pour 'Age' selon le statut 'Exited'
def calculate_percentile(series, percentile):
    return series.quantile(percentile / 100)

stats = {
    'Exited': [],
    'Q1': [],
    'Q2': [],
    'Médiane': [],
    'Q3': [],
    'Q99': [],
    'Moyenne': [],
    'Écart-type': []
}

for exited_status in [0, 1]:
    subset = data[data['Exited'] == exited_status]['Age']
    stats['Exited'].append(exited_status)
    stats['Q1'].append(calculate_percentile(subset, 25))
    stats['Q2'].append(calculate_percentile(subset, 50))
    stats['Médiane'].append(subset.median())
    stats['Q3'].append(calculate_percentile(subset, 75))
    stats['Q99'].append(calculate_percentile(subset, 99))
    stats['Moyenne'].append(subset.mean())
    stats['Écart-type'].append(subset.std())

# Ajout des statistiques globales (pour tous les clients)
all_ages = data['Age']
stats['Exited'].append('All')
stats['Q1'].append(calculate_percentile(all_ages, 25))
stats['Q2'].append(calculate_percentile(all_ages, 50))
stats['Médiane'].append(all_ages.median())
stats['Q3'].append(calculate_percentile(all_ages, 75))
stats['Q99'].append(calculate_percentile(all_ages, 99))
stats['Moyenne'].append(all_ages.mean())
stats['Écart-type'].append(all_ages.std())

# Création d'un DataFrame pour afficher les résultats sous forme de tableau
age_stats_df = pd.DataFrame(stats)

# Affichage du tableau
print(age_stats_df)
```

# Tableau de la variable balance

```python
import pandas as pd
import numpy as np

# Chargement des données
file_path = 'Documents\\Churn_Modelling.csv'
data = pd.read_csv(file_path)

# Fonction pour calculer Q99
def quantile_99(series):
    return np.percentile(series, 99)

# Création du tableau récapitulatif
summary_table = data.groupby('Exited').agg(
    Q1=('Balance', lambda x: np.percentile(x, 25)),
    Q2=('Balance', lambda x: np.percentile(x, 50)),
    Médiane=('Balance', lambda x: np.median(x)),
    Q3=('Balance', lambda x: np.percentile(x, 75)),
    Q99=('Balance', quantile_99),
    Moyenne=('Balance', 'mean'),
    Écart_type=('Balance', 'std')
).reset_index()

# Affichage du tableau
print(summary_table)

# Sauvegarde du tableau au format CSV si nécessaire
summary_table.to_csv('/mnt/data/summary_table_balance.csv', index=False)
```

# Aalyse bivariee des variables quantitavies avec la variable cible EXITED (courbe boxplot et courbe de densite) 

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Chargement des données
file_path = 'Documents\\churn_Modelling.csv'
data = pd.read_csv(file_path)

# Exclusion des colonnes 'CustomerId' et 'Surname'
data = data.drop(['CustomerId', 'Surname'], axis=1)

# Identification des variables quantitatives
quantitative_vars = data.select_dtypes(include=['float64', 'int64']).columns.drop('Exited')

# Initialisation d'une liste pour stocker les statistiques descriptives
stats_summary = []

# Boucle pour l'analyse bivariée entre 'Exited' et chaque variable quantitative
for var in quantitative_vars:
    # Calcul des statistiques descriptives pour Exited = 0 et Exited = 1
    stats_0 = data[data['Exited'] == 0][var].describe()
    stats_1 = data[data['Exited'] == 1][var].describe()
    
    # Ajout des statistiques au tableau récapitulatif
    stats_summary.append({
        'Variable': var,
        'Moyenne (Exited=0)': stats_0['mean'],
        'Moyenne (Exited=1)': stats_1['mean'],
        'Médiane (Exited=0)': stats_0['50%'],
        'Médiane (Exited=1)': stats_1['50%'],
        'Écart-type (Exited=0)': stats_0['std'],
        'Écart-type (Exited=1)': stats_1['std']
    })
    
    # Création des boxplots
    plt.figure(figsize=(10, 5))
    sns.boxplot(x='Exited', y=var, data=data, palette="Set3")
    plt.title(f"Boxplot de '{var}' selon Exited")
    plt.xlabel('Exited')
    plt.ylabel(var)
    plt.show()
    
    # Création des courbes de densité
    plt.figure(figsize=(10, 5))
    sns.kdeplot(data[data['Exited'] == 0][var], label='Exited = 0', shade=True, color="blue")
    sns.kdeplot(data[data['Exited'] == 1][var], label='Exited = 1', shade=True, color="red")
    plt.title(f"Courbes de densité de '{var}' selon Exited")
    plt.xlabel(var)
    plt.ylabel("Densité")
    plt.legend()
    plt.show()

# Création d'un DataFrame avec le tableau récapitulatif des statistiques
stats_summary_df = pd.DataFrame(stats_summary)

# Affichage du tableau des résultats
print("\nRésumé des caractéristiques des variables quantitatives selon 'Exited':")
print(stats_summary_df)
```
